```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

task=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
task_1=task.copy()
task_11=task.copy()

#convert gender to category
def gender_category( gender):
    if gender=='Female':
        return 1
    else:
        return 0
task_11['Gen']=task_11['gender'].apply(gender_category)    
task_11.drop(axis=1,columns='gender',inplace=True)



#convert partner to category
def Partner_category(Partner):

    if Partner=='Yes':
        return 1
    else:
        return 0
task_11['partner']=task_11['Partner'].apply(Partner_category)    
task_11.drop(axis=1,columns='Partner',inplace=True)         
          

#convert Dependents to category
def Dependents_category(Dependents):
    if Dependents=='Yes':
        return 1
    else:
        return 0
task_11['Dep.']=task_11['Dependents'].apply(Dependents_category)    
task_11.drop(axis=1,columns='Dependents',inplace=True)




##   
def phone_ser_category(PhoneService):
    if PhoneService=='Yes':
        return 1
    else:
         return np.nan 
        
task_11['Phone']=task_11['PhoneService'].apply(phone_ser_category)    
task_11.drop(axis=1,columns='PhoneService',inplace=True)
task_11['Phone'].dropna(inplace=True)

##
def Multiline_category(MultipleLines):
    if MultipleLines =='Yes':
        return 1
    elif MultipleLines=='No':
        return 0
    else:
         return np.nan     
task_11['Multiphone']=task_11['MultipleLines'].apply(Multiline_category)    
task_11.drop(axis=1,columns='MultipleLines',inplace=True)
task_11['Multiphone'].dropna(inplace=True)


##
def Internet_category(InternetService):
    if InternetService =="DSL":   #DSL 
        return 1
    elif InternetService=="Fiber optic":  #Fiber optic
        return 0
    else:
         return np.nan     
task_11['Internet']=task_11['InternetService'].apply(Internet_category)    
task_11.drop(axis=1,columns='InternetService',inplace=True)
task_11['Internet'].dropna(inplace=True)
#
def O_security_category(OnlineSecurity):
    if OnlineSecurity=='Yes':
        return 1
    elif OnlineSecurity=='No':
        return 0
    else:
         return np.nan     
task_11['O.Security']=task_11['OnlineSecurity'].apply(O_security_category)    
task_11.drop(axis=1,columns='OnlineSecurity',inplace=True)
task_11['O.Security'].dropna(inplace=True)


#
def O_Backup_category(OnlineBackup):
    if OnlineBackup=='Yes':
        return 1
    elif OnlineBackup=='No':
        return 0
    else:
         return np.nan     
task_11['O.Backup']=task_11['OnlineBackup'].apply(O_Backup_category)    
task_11.drop(axis=1,columns='OnlineBackup',inplace=True)
task_11['O.Backup'].dropna(inplace=True)


#
def DevProtection_category(DeviceProtection):
    if DeviceProtection=='Yes':
        return 1
    elif DeviceProtection=='No':
        return 0
    else:
         return np.nan     
task_11['DevProtection']=task_11['DeviceProtection'].apply(DevProtection_category)    
task_11.drop(axis=1,columns='DeviceProtection',inplace=True)
task_11['DevProtection'].dropna(inplace=True)

##
def TechSupport_category(TechSupport):
    if TechSupport=='Yes':
        return 1
    elif TechSupport=='No':
        return 0
    else:
         return np.nan     
task_11['TechSup.']=task_11['TechSupport'].apply(TechSupport_category)    
task_11.drop(axis=1,columns='TechSupport',inplace=True)
task_11['TechSup.'].dropna(inplace=True)


##
def stream_tv_category(StreamingTV):
    if StreamingTV=='Yes':
        return 1
    elif StreamingTV=='No':
        return 0
    else:
         return np.nan     
task_11['Stream Tv.']=task_11['StreamingTV'].apply(stream_tv_category)    
task_11.drop(axis=1,columns='StreamingTV',inplace=True)
task_11['Stream Tv.'].dropna(inplace=True)



##
def stream_movies_category(StreamingMovies):
    if StreamingMovies =='Yes':
        return 1
    elif StreamingMovies=='No':
        return 0
    else:
         return np.nan     
task_11['Stream Movies.']=task_11['StreamingMovies'].apply(stream_movies_category)    
task_11.drop(axis=1,columns='StreamingMovies',inplace=True)
task_11['Stream Movies.'].dropna(inplace=True)



###
def Paper_less_Billing_category(PaperlessBilling):
    if PaperlessBilling =='Yes':
        return 1
    elif PaperlessBilling=='No':
        return 0
    else:
         return np.nan     
task_11['PaperBilling.']=task_11['PaperlessBilling'].apply(Paper_less_Billing_category)    
task_11.drop(axis=1,columns='PaperlessBilling',inplace=True) 
task_11['PaperBilling.'].dropna(inplace=True)


###
def Payment_Method_category(PaymentMethod):
    if PaymentMethod =='Bank transfer (automatic)':
        return 1.1
    elif  PaymentMethod =='Credit card (automatic)':
        return 1.2
    elif  PaymentMethod == 'Electronic check':
        return 2.1
    elif  PaymentMethod =='Mailed check':
        return 2.2
    else:
         return np.nan     
task_11['Payment']=task_11['PaymentMethod'].apply(Payment_Method_category)    
task_11.drop(axis=1,columns='PaymentMethod',inplace=True) 
task_11['Payment'].dropna(inplace=True)

###########
def Churn_category(Churn):
    if Churn=='Yes':
        return 1
    else:
         return np.nan 
        
        
task_11['Churn_1']=task_11['Churn'].apply(Churn_category)    
task_11.drop(axis=1,columns='Churn',inplace=True)
task_11['Churn_1'].dropna(inplace=True)

###########
def Co_category(Contract):
    if Contract=='Month-to-month':
        return 0
    elif Contract=='One year':
        return 1
    elif Contract=='Two year':
        return 2
    
    else:
         return np.nan 
        
        
task_11['Con.']=task_11['Contract'].apply(Co_category)    
task_11.drop(axis=1,columns='Contract',inplace=True)
task_11['Con.'].dropna(inplace=True)



##########

task_11.dropna(inplace=True)
task_11.shape
```
